```python
import pandas as pd
df=pd.read_csv('/content/pandas practice_dataset.csv')
```

```python
df.head(5)
```

**Find the average salary by Department**

```python
df.groupby('Department')['Salary'].mean()
```

**Find the maximum salary by Department.**

```python
df.groupby('Department')['Salary'].max()
```

**Count the number of employees per Department**

```python
df['Department'].value_counts()
```

```python
df.groupby('Department').size()['IT'] #count particular one department
```

```python
df.groupby(['Department','Gender'])['Rating'].mean()
```
