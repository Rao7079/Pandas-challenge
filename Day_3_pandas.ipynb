```python
import pandas as pd
df=pd.read_csv('/content/pandas practice_dataset.csv')
```

```python
df.head(10)
```

**Fill missing salaries with the average salary.**

```python
df['Salary'].fillna(df['Salary'].mean(), inplace=True)
print(df)
```

```python
df['Salary'] = pd.to_numeric(df['Salary'], errors='coerce')
```

**Fill missing cities with "Unknown"**

```python
df['City'].fillna("Unkown",inplace=True)
df.head(20)
```

**Remove duplicate rows.**

```python
df.drop_duplicates(inplace=True)
df
```

**Rename the column Salary to Salary_Amount.**

```python
df.rename(columns={'Salary':'Salary_Amaount'},inplace=True)
df.head(1)
```

**Strip extra spaces and convert all city names to lowercase.**

```python
df['City']=df['City'].str.strip().str.lower()
df.head(5)
```
