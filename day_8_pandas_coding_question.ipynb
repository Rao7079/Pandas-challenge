```python
import pandas as pd
df=pd.read_csv('/content/10 coding question.csv')
```

**Load CSV and check shape, data types, summary**

```python
check_shape=df.shape
print(check_shape)
check_type=df.dtypes
print(check_type)

check_summary=df.info()
print(check_summary)
```

**Filter employees having Salary > 50000 and Department = "IT"**

```python
df[(df['Salary']>50000) & (df['Department']=='IT')]
```

Create a new column “Annual_Salary” = Salary * **12**

```python
df['Annual_Salary']=df['Salary']*12
df
```

Replace all missing values in “Age” column with median

```python
df['Age'].fillna(df['Age'].median())
```

**Group employees and find average salary by Department**

```python
df.groupby('Department')['Salary'].mean()
```

**Find top 3 highest paid employees**

```python
df.sort_values(by='Salary',ascending=False).head(3)
```

**how to find total missing value in your dataframe**

```python
df.isnull().sum()
```

**Create a pivot table: mean Salary by Department vs Gender**

```python
df.pivot_table(
    index='Department',
    columns='Gender',
    values='Salary',
    aggfunc='mean'

)
```

**Delete duplicate rows**

```python
df.drop_duplicates(inplace=True)
df
```

**Convert “Join_Date” column to datetime and extract year**

```python
df['Join_Date']=pd.to_datetime(df['Join_Date'])
df['year']=df['Join_Date'].dt.year
df
```
